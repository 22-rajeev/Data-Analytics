```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])
y = np.array([1,3,5,7,9])

plt.scatter(x,y)
```

```python
# Now we need to add up some synthetic data points between these known data points.
# Hence we will use linear interpolation over here since the data is linear in this case.
```

```python
x_new = np.linspace(1,5,10) # creates new vales of x from 1 to 5 
y_interpolated = np.interp(x_new , x, y)
```

```python
plt.scatter(x_new , y_interpolated)
```

```python
# The above data has now more points as we used interpolation technique to synthetically add up more data between known data points
```

```python
# Cubic Interpolation:

x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])

plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
f = interp1d(x,y, kind='cubic') # f is a cubic interpolation object from scipy 
f
```

```python
x_new = np.linspace(1,5,10)
y_interp = f(x_new)
```

```python
y_interp
```

```python
 plt.scatter(x_new , y_interp)
```

```python
 #polynomial interpolation 
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,1,25])

p = np.polyfit(x,y,2)
```

```python
x_new = np.linspace(x,y,10)
y_interp = np.polyval(p,x_new) 

plt.scatter(x,y)
```

```python
plt.scatter(x_new , y_interp)
```
